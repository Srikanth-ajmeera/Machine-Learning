**Categorical Columns**

```python
categorical_columns = ['City']
df_cleaned = df.dropna()
for col in categorical_columns:
    print(f"Frequency count for {col}:\n", df_cleaned[col].value_counts())
for col in categorical_columns:
    plt.figure(figsize=(16, 13))
    sns.countplot(y=col, data=df_cleaned)
    plt.title(f"Bar Plot of {col}")
    plt.show()
```

The bar plot shows the distribution of cities in the dataset. The most frequent city is Bengaluru, followed by Hyderabad and Pune. The distribution is skewed to the right, with a longer tail on the less frequent city side. This suggests that there are some cities with significantly fewer instances compared to the most frequent ones.

## **2.Multivariate Analysis**

```python
df['Average Fees'] = pd.to_numeric(df['Average Fees'], errors='coerce')
df['Established Year'] = pd.to_numeric(df['Established Year'], errors='coerce')

correlation_matrix = df[['Established Year', 'Average Fees']].corr()
correlation_matrix
```

```python
df_cleaned = df.dropna()
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Average Fees', y='Rating', data=df_cleaned)
plt.title('Average Fees vs Rating')
```

The scatter plot visualizes the relationship between average fees and rating.

**key observations:**

**No Strong Correlation:**There appears to be no strong linear correlation between average fees and rating. The points are scattered across the plot, indicating that there is no clear trend of increasing or decreasing rating with higher fees.

**Clustering:**There might be some clustering of points in certain regions, suggesting that institutions with similar average fees tend to have ratings within a specific range.

**Outliers:**A few outliers can be observed, indicating institutions with either very high fees and relatively low ratings, or vice versa.

```python
df_cleaned = df.dropna()
plt.figure(figsize=(4, 6))
sns.boxplot(x='College Type', y='Average Fees', data=df_cleaned)
plt.title('Average Fees by College Type (Private vs Government)')
```

The box plot visualizes the distribution of average fees for private and public/government colleges.
**key observations:**

**Median Fees:** The median average fees for private colleges are slightly higher than those for public/government colleges.

**IQR:** The interquartile range (IQR) for private colleges is wider, indicating a greater spread of fees among private institutions.

**Outliers:** There are outliers present in the private college category, suggesting that some private colleges have significantly higher average fees compared to the majority.

**Distribution:**The distribution of fees for both private and public/government colleges appears to be skewed to the right, with a longer tail on the higher fee side.

```python
numerical_columns = ['Total Student Enrollments', 'Total Faculty', 'Established Year', 'Rating', 'Average Fees']
for column in numerical_columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')
df_cleaned = df[numerical_columns].dropna()
correlation_matrix = df_cleaned.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
```

The provided correlation heatmap visualizes the relationships between numerical features in a dataset.

**key observations:**

**Strong Positive Correlation:**
Total Student Enrollments and Total Faculty: A strong positive correlation exists between these features, indicating that institutions with higher student enrollments tend to have more faculty members.

**Moderate Positive Correlation:**
Total Student Enrollments and Rating: A moderate positive correlation suggests that institutions with higher enrollments generally have better ratings.

**Weak or No Correlation:**
Established Year and other features: There is little or no correlation between the established year of an institution and the other numerical features, suggesting that age does not significantly impact factors like student enrollment, faculty size, or average fees.

**Negative Correlation:**
Average Fees and Total Faculty: A weak negative correlation indicates that institutions with higher average fees might have slightly lower faculty-to-student ratios.
